```python
import pandas as pd
import re
df=pd.read_csv('Data/Attachment ambiguity WITH.csv',encoding='latin-1',header=None)
df
```

```python
x=list(df[0])
x
```

```python
import string
R_patterns = [(r'won\'t','will not'),
             (r'can\'t','cannot'),
             (r'i\'m','i am'),
             (r'(\w+)\'ll','\g<1> will'),
             (r'(\w+)n\'t','\g<1> not'),
             (r'(\w+)\'ve','\g<1> have'),
             (r'(\w+)\'re','\g<1> are'),
              (r'(\w+)\'it\'s','\g<1> it is'),
             ]
class REReplacer(object):
    def __init__(self,patterns=R_patterns):
        self.patterns = [(re.compile(regex),rep1) for (regex,rep1) in patterns]
    def replace(self,text):
        s=text
        for(pattern,rep1) in self.patterns:
            s=re.sub(pattern,rep1,s)
        return s
rep_word=REReplacer()
x1=[]
for i in x:
    n=rep_word.replace(i)
    n=n.translate(str.maketrans("","",string.punctuation))
    x1.append(n.lower())
```

```python
import nltk
from nltk.corpus import stopwords
words=[]
for i in x1:
    words.append(nltk.word_tokenize(i))
words
```

```python
stop_words=set(stopwords.words('english'))
```

```python
token=[]
for i in words:
    for j in i:
        token.append(j.lower())
st=set()
for i in token:
    st.add(i)
```

```python
st
```

```python
count=dict()
for i in st:
    cnt=0
    for j in token:
        if(i==j):
            cnt+=1
    count[i]=cnt
```

```python
for k,v in count.items():
    print(k,v)
```

```python
from nltk.util import bigrams
bi_grams=list(bigrams(token))
```

```python
noun = input("Enter the Noun: ")
verb = input("Enter the Verb: ")
prep = input("Enter the Preposition: ")
```

```python
bigram_counts = {}
for bigram in bi_grams:
    bigram_counts[bigram] = bigram_counts.get(bigram, 0) + 1
```

```python
for i,j in bigram_counts.items():
    print(i,j)
```

```python
n = count.get(noun.lower(), 0)
v = count.get(verb.lower(), 0)
p_n = bigram_counts.get((noun.lower(), prep.lower()), 0)
p_v = bigram_counts.get((verb.lower(), prep.lower()), 0)
print(f"Frequency of Noun: {n}\nFrequency of Verb: {v}\nFrequency of P and N: {p_n}\nFrequency of P and V: {p_v}")
```

```python
import math
def cal_prob(p_v,p_n,v,n):
    prob_v = p_v / v if p_v and v > 0 else 0.9
    prob_n = p_n / n if p_n and n > 0 else 0.9
    return prob_v,prob_n
prob_v,prob_n=cal_prob(p_v,p_n,v,n)
print(prob_v,prob_n)
def cal_lam(prob_v,prob_n):
    _lambda = (prob_v * (1 - prob_n)) / prob_n
    _lambda=math.log2(_lambda)
    return _lambda
_lambda=cal_lam(prob_v,prob_n)
print(_lambda)
# Determine the attachment of the preposition
if _lambda > 0:
    print("The Preposition is attached with Verb.")
    
elif _lambda < 0:
    print("The Preposition is attached with Noun.")
else:
    print("The Preposition attachment cannot be determined.")
```
